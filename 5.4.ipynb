```python
from random import randint
import random
correct=0
wrong=0
vict={'Россия': 'Москва','Англия': 'Лондон','Япония': 'Токио'}
num=int(input("введите количество стран и столиц которое хотите добавить в словарь: "))
for i in range(0,num,1):
  country=input("введите страну ")
  capital=input("введите столицу ")
  vict[country] = capital
  

attempts=int(input("Введите количество желаемых попыток "))

while attempts>0:
  x=randint(0, len(vict))
  random_country=list(vict.keys())[x]
  random_capital=list(vict.values())[x]
  print("Угадайте столицу страны",random_country)
  user_guess=input()
  if user_guess==random_capital:
    correct=correct+1
    attempts=attempts-1
  else:
   wrong=wrong+1
   attempts=attempts-1

print("Количество правильных ответов: ",correct)
print("Количество неправильных ответов: ",wrong)

```

```python
vowels = ['а', 'о', 'у', 'э', 'ы', 'я', 'ё', 'е', 'ю', 'и']
def vowel(s): 
  k = 0
  for element in s:
    if element in vowels:
      k=k+1
  return k
def consonant(s):
  k = 0
  for element in s:
    if element not in vowels:
      k=k+1
  return k
s = input()
print(vowel(s))
print(consonant(s)) 
```
